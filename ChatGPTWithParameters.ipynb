```python
!pip install openai # Installs openai library

import openai # Imports openai library to access ChatGPT

# PASTE YOUR KEY IN THE QUOTATION MARKS BELOW
apikey = "" # Defines API key, which is needed to use ChatGPT

conversation = [] # Variable that allows ChatGPT to remember previous prompts

def chatgpt(prompt, conversation = conversation, key = apikey, 
            max_tokens = 1000, # Maximum length the answer can be
            temperature = 0.5, # How "creative" or original the answer will be
            frequency_penalty = 0): # How much the AI will avoid repeating phrases
      # More specifics on the parameters can be found in the README file

  # Starts ChatGPT with API Key and gives it prompt
  openai.api_key = key
  completion = openai.ChatCompletion.create(model = "gpt-3.5-turbo", 
    messages = conversation + [{"role":"user", "content":prompt}], 
    max_tokens = max_tokens,
    temperature = temperature,
    frequency_penalty = frequency_penalty)
  
  # Displays prompt and response
  print('Prompt:', prompt)
  print('\nResponse:')
  print(completion.choices[0]['message']['content'])

  # Updates conversation
  conversation += [{"role":"user", "content":prompt}, completion.choices[0]['message']]
```

```python
chatgpt("")
```
