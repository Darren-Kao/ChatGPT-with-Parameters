```python
!pip install openai # Installs openai library

import openai # Imports openai library to access ChatGPT

# PASTE YOUR KEY IN THE QUOTATION MARKS BELOW
apikey = "" # Defines API key, which is needed to use ChatGPT

conversation = [] # Variable that allows ChatGPT to remember previous prompts

def chatgpt(prompt, conversation = conversation, key = apikey, 
            suffix = None, 
            max_tokens = 100, # Maximum length the response can be
            temperature = 0.5, # How "creative" or original the response will be
            top_p = 1, # How likely it is for the response to contain words/phrases with lesser probabilities (like temperature)
            stream = False, # Whether response will be sent from server all at once or token by token
            n = 1, # How many responses the program will return
            logprobs = None, # How many most likely tokens the program will give log probabilities for
            stop = None, # Phrases that, if reached, will end the response
            prescence_penalty = 0, # How much the AI will try to say new phrases
            frequency_penalty = 0, # How much the AI will avoid repeating phrases (like prescence penalty)
            best_of = 1, # How many responses the AI will generate before returning the n responses with highest token probabilities
            logit_bias = None): # A way to modify how likely certain tokens are to appear in the response
  
  # Some commonly-used parameters are explained in the corresponding YouTube video
  # More specifics on the parameters can be found on the official API documentation, linked below

  # Starts ChatGPT with API Key and gives it prompt
  openai.api_key = key
  completion = openai.ChatCompletion.create(model = "gpt-3.5-turbo", 
    messages = conversation + [{"role":"user", "content":prompt}], 
    suffix = suffix, max_tokens = max_tokens, 
    temperature = temperature, top_p = top_p, 
    stream = stream, n = n, logprobs = logprobs, 
    stop = stop, prescence_penalty = prescence_penalty, 
    frequency_penalty = frequency_penalty, 
    best_of = best_of, logit_bias = logit_bias)
  
  # Displays prompt and response
  print('Prompt:', prompt)
  print('\nResponse:')
  print(completion.choices[0]['message']['content'])

  # Updates conversation
  conversation += [{"role":"user", "content":prompt}, completion.choices[0]['message']]
```

```python
chatgpt("")
```
